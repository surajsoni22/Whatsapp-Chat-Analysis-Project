```python
import re
import pandas as pd
```

```python
f = open("chat.txt",'r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
#doing this will delete all the date patern and shows only message

messages = re.split(pattern,data)[1:]  
```

```python
messages
```

```python
#it shows only date
dates = re.findall(pattern,data) 
```

```python
dates
```

```python
df = pd.DataFrame({"user_message":messages,"message_date":dates})
```

```python
df
```

```python
df['message_date']
```

```python
#changing message date into date time format

df["message_date"] = pd.to_datetime(df['message_date'], format = "%d/%m/%y, %H:%M - ") 
```

```python
df['message_date']
```

```python
df.shape
```

```python
df
```

```python
#spliting the user message and seperating the name and message 

df.rename(columns={'message_date':'date'}, inplace = True)
```

```python
df
```

```python
users = []
messages= []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group notification')
        messages.append(entry[0])
        
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'],inplace=True)
```

```python
periods = []
for hour in df[['day_name','hour']]['hour']:
    if hour ==23:
        periods.append(str(hour) + "-" +str("00"))
    elif hour ==0:
        periods.append(str("00") + "-" +str(hour+1))
    else:
        periods.append(str(hour) + "-" +str(hour+1))
```

```python
df['period'] = periods
```

```python
df
```

```python
import seaborn as sns
plt.figure(figsize = (20,6))
sns.heatmap(df.pivot_table(index = 'day_name',columns = 'period',values='message' ,aggfunc = 'count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
